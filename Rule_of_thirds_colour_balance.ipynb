```python
def colour_balance_ver(name):
  image_BGR = cv2.imread(name, cv2.IMREAD_UNCHANGED)
  sum = 0
  for i in range(image_BGR.shape[0]):
    for j in range(int(image_BGR.shape[1]/2)):
      cur_sum = 0
      cur_sum += (image_BGR[i][j][0] - image_BGR[i][image_BGR.shape[1]-1-j][0]) ** 2
      # print((image_BGR[i][j][0] - image_BGR[i][image_BGR.shape[1]-1-j][0]) ** 2)
      cur_sum += (image_BGR[i][j][1] - image_BGR[i][image_BGR.shape[1]-1-j][1]) ** 2
      cur_sum += (image_BGR[i][j][2] - image_BGR[i][image_BGR.shape[1]-1-j][2]) ** 2
      if image_BGR.shape[2] == 4:
        if image_BGR[i][j][3] == 0 or image_BGR[i][image_BGR.shape[1]-1-j][3] == 0:
          continue
      sum += cur_sum / float(image_BGR.shape[0] * image_BGR.shape[1])
  return sum
```

```python
def colour_balance_horiz(name):
  image_BGR = cv2.imread(name, cv2.IMREAD_UNCHANGED)
  sum = 0
  for i in range(int(image_BGR.shape[0]/2)-100):
    for j in range(image_BGR.shape[1]):
      cur_sum = 0
      cur_sum += (image_BGR[i+200][j][0] - image_BGR[image_BGR.shape[0]-1-i][j][0]) ** 2
      # print((image_BGR[i][j][0] - image_BGR[i][image_BGR.shape[1]-1-j][0]) ** 2)
      cur_sum += (image_BGR[i+200][j][1] - image_BGR[image_BGR.shape[1]-1-i][j][1]) ** 2
      cur_sum += (image_BGR[i+200][j][2] - image_BGR[image_BGR.shape[1]-1-i][j][2]) ** 2
      if image_BGR.shape[2] == 4:
        cur_sum *= image_BGR[i+200][j][3] * image_BGR[image_BGR.shape[1]-1-i][j][3]
      sum += cur_sum / float(image_BGR.shape[0] * image_BGR.shape[1])
  return sum
```
